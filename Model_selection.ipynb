```python
import pandas as pd
```

```python
# load dataframe mvp1
mvp1_path = "web-dash//mvp1data.csv"
df = pd.read_csv(mvp1_path)
df
```

Drop 100 rows for final testing

```python
mvp2_path = "web-dash//mvp2data.csv"
df_100_test = pd.read_csv(mvp2_path)

# Identify rows in df_b that are not in df_a based on "_unit_id"
mask = ~df['_unit_id'].isin(df_100_test['_unit_id'])

# Filter df_b to keep rows not in df_a
df = df[mask]
```

```python
df
```

Find duplicate row

```python
df.duplicated(subset=['sentence_1', 'sentence_2'], keep=False)
```

```python
x = []
y = []

# Get the maximum value of "confidence:stddev" for each group of duplicate rows
df_sorted = df.sort_values('confidence:stddev', ascending=False)
df_final = df_sorted.drop_duplicates(subset=['sentence_1', 'sentence_2'], keep='first').reset_index(drop=True)
df_final = df_final[['sentence_1', 'sentence_2', 'confidence:stddev', 'relation_BAF']]

# Loop through each row in the DataFrame
for i in range(len(df_final['sentence_1'])):
    sentence = []
    sentence.append(df_final["sentence_1"][i])
    sentence.append(df_final["sentence_2"][i])
    x.append(sentence)
    y.append(df_final["relation_BAF"][i])
```

```python
df_final
```

```python
Check = list(df_final.duplicated(subset=['sentence_1', 'sentence_2'], keep=False))
for i in Check:
    if i == True:
        print("Duplicated")
print('end checking')
```

```python
print(len(x))
print(len(y))
```

Selected train test dataset

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)

# Displaying the shapes of the resulting datasets
print("X_train shape:", len(X_train))
print("X_test shape:", len(X_test))
print("y_train shape:", len(y_train))
print("y_test shape:", len(y_test))
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

# Creating a TF-IDF vectorizer
tfidf_vectorizer = TfidfVectorizer()

# Transforming the entire training set
X_train_tfidf = tfidf_vectorizer.fit_transform([sentence[0] + ' ' + sentence[1] for sentence in X_train])

# Transforming the entire testing set using the same vectorizer
X_test_tfidf = tfidf_vectorizer.transform([sentence[0] + ' ' + sentence[1] for sentence in X_test])
```

Apply SVM Algorithm

```python
from sklearn import svm
```

```python
clf_SVC = svm.SVC()
clf_SVC.fit(X_train_tfidf, y_train)
```

```python
# Evaluating the model
accuracy_SVC = clf_SVC.score(X_test_tfidf, y_test)
print("Accuracy:", accuracy_SVC)
```

Apply DecisionTree Algorithm

```python
from sklearn import tree
```

```python
clf_tree = tree.DecisionTreeClassifier()
clf_tree = clf_tree.fit(X_train_tfidf, y_train)
```

```python
accuracy_tree = clf_tree.score(X_test_tfidf, y_test)
print("Accuracy:", accuracy_tree)
```

Apply Naive Bayes Algorithm

```python
from sklearn.naive_bayes import GaussianNB
```

```python
gnb = GaussianNB()
gnb = gnb.fit(X_train_tfidf.toarray(), y_train)
```

```python
accuracy_gnb = gnb.score(X_test_tfidf.toarray(), y_test)
print("Accuracy:", accuracy_gnb)
```

Apply LogisticRegression Algorithm 

```python
from sklearn.linear_model import LogisticRegression
```

```python
clf_LR = LogisticRegression(random_state=0)
clf_LR = clf_LR.fit(X_train_tfidf, y_train)
```

```python
accuracy_LR = clf_LR.score(X_test_tfidf, y_test)
print("Accuracy:", accuracy_LR)
```

Table Result

```python
print("Accuracy of SVM:", accuracy_SVC)
print("Accuracy of Decision Tree:", accuracy_tree)
print("Accuracy of NB:", accuracy_gnb)
print("Accuracy of LR:", accuracy_LR)
```

```python
#list a topics name
group_marked = df_100_test.groupby('topic')
list_topic = []
for topic, group in group_marked:
    list_topic.append(topic)


def find_topic(topic, model):
    senten = []
    rela = []
    selected_rows = df_100_test[df_100_test['topic'].str.contains(topic, case=False)]
    ind_1 = list(selected_rows['sen1_inx'])
    ind_2 = list(selected_rows['sen2_inx'])
    for i in range(len(selected_rows)):
        # Use the correct column names based on your DataFrame
        if ind_1[i] not in senten: #check is there any sen1 in senten or not to prevent duplicated
            senten.append(ind_1[i])
        if ind_2[i] not in senten:
            senten.append(ind_2[i])
        if df_100_test[model][i]:
            rela.append(f"{ind_1[i]},{ind_2[i]}")
    return senten, rela


def convert_into_input(sen, rel, filename):
    #convert senten and rela into input file format
    ipl =""
    for i in sen:
        ipl = ipl+f"arg({i})."
    for i in rel:
        ipl = ipl+f"att({i})."
    with open(filename, 'w') as file:
        file.write(ipl)

def loop_topic(model):
    for topic in list_topic:
        sentence, relation = find_topic(topic, model)
        filename = f"ASP_input//{model}//input_{topic}.pl"
        filename = filename.replace(" ", "_")
        convert_into_input(sentence, relation, filename)
```

```python
def markcsv(number, model):
    df_100_test[f'satis_{model}'] = 0
    for i in range(len(df_100_test["sen1_inx"])):
        if df_100_test["sen1_inx"][i] in number:
            df_100_test[f'satis_{model}'][i] = 1

def readline(file_name, model):
    file1 = open(file_name, 'r')
    Lines = file1.readlines()
    count = 0
    numbers = []
    # Strips the newline character
    for line in Lines:
        count += 1
        word = line.split(" ")
        for i in word:
            if "in(" in i:
                number = i[3:len(i)-1]
                # print(number)
                numbers.append(int(number))
    markcsv(numbers, model)
    return numbers

        # print("Line{}: {}".format(count, line.strip()))
```

### Accuracy in each model
#### Large Language Models (Text Davin cii 003, GPT3.5 turbo, and gpt 3.5 turbo version 0613)


```python
import matplotlib.pyplot as plt
from sklearn.metrics import classification_report
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay
```

```python
y_true = list(df_100_test["relation_BAF"])
def confu_max(y_true, y_preds):
    target_names = ['0', '1', '2']
    print(classification_report(y_true, y_preds, target_names=target_names, zero_division=1))
    cm = confusion_matrix(y_true, y_preds)
    disp = ConfusionMatrixDisplay(confusion_matrix=cm)
    disp.plot()
    plt.show()
```

```python
y_davin = list(df_100_test["gpt_davinci"])
y_gpt35 = list(df_100_test["gpt_35"])
y_gpt35_0613 = list(df_100_test["gpt_35_0613"])
```

```python
print("original vs davin")
print(len(y_true),len(y_davin))
confu_max(y_true, y_davin)
print("original vs gpt 3.5")
confu_max(y_true, y_gpt35)
print("original vs gpt 3.5 version 0613")
confu_max(y_true, y_gpt35_0613)
```
